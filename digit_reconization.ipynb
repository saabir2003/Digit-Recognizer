```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
import tensorflow as tf
```

```python
data=pd.read_csv('train.csv')
```

```python
data.describe()
```

```python
data.head()
```

```python
X_train=data.drop(columns=['label'])
y_train=data['label']
```

```python
print(X_train)
print(y_train)
```

```python
tf.random.set_seed(32)
```

```python
#normalizing data
X_train=X_train/255
```

```python
X_train.describe()
```

```python
X_train
```

```python
from google.colab import files
X_train.to_csv("X_train.csv")
files.download('X_train.csv')
```

```python
tf.random.set_seed(32)
```

```python
model1=tf.keras.Sequential([
    tf.keras.layers.Dense(100,activation="relu"),
    tf.keras.layers.Dense(128,activation='relu'),
    tf.keras.layers.Dense(10,activation='softmax')]
)

model1.compile(optimizer='Adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

```python
history1=model1.fit(X_train,y_train,epochs=25)
```

```python
test=pd.read_csv('test.csv')
test.head()
```

```python
test.drop(index=0)
test
```

```python
predict=model1.predict(test)
```

```python
predict.shape
```

```python
predict=predict.round()
```

```python
predict = tf.cast(predict, dtype=tf.int32)
```

```python
j=1
l=[]
for i in predict:
  print(i)
  indexes = tf.where(tf.equal(i, 1))
  print(indexes.numpy().squeeze())
  l.append([j,indexes.numpy().squeeze()])
  j=j+1
```

```python
exit=pd.DataFrame(l)
col=["ImageId","Label"]
exit.columns=col
exit.head()
```

```python

from google.colab import files
exit.to_csv('sub.csv')
files.download('sub.csv')
```
